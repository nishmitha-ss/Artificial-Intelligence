```python
import scipy.io
import numpy as np

data = scipy.io.loadmat('data.mat')
train_X = data['X']
train_y = data['y']
test_X = data['Xval']
test_y = data['yval']
```

```python
#Neural network
from keras import models
from keras import layers
from keras.layers import Dense
from keras import losses
from keras import metrics
from keras.models import Sequential
from keras import optimizers

model = models.Sequential()
model.add(Dense(16, activation='relu',input_dim=2))
model.add(Dense(16,activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(train_X, train_y,epochs=200,verbose=2,validation_data=(test_X,test_y))
```

```python
#Testing accuracy and loss of neural network

score = model.evaluate(test_X, test_y, verbose = 0)
print("Test score: ", score[0])
print("Test accuracy: ", score[1])
```

```python
#Scatter plot for traiing datasets
import matplotlib.pyplot as plt

plt.scatter(train_X.T[0, :], train_X.T[1, :], c=train_y.T[0, :], s=40, cmap=plt.cm.Spectral);
```

```python
#Desicion boundary for trainng datasets
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

svm = SVC(C=0.5, kernel='linear')
svm.fit(train_X, train_y.flatten())
plot_decision_regions(train_X, train_y.flatten(), clf=svm, legend=2);
plt.xlim(-0.6, 0.4)
plt.ylim(-0.8, 0.8)
```

```python
regur=model.predict(test_X)
```

```python
#Desicion boundary for test datasets
svm = SVC(C=0.5, kernel='linear')
svm.fit(test_X, test_y.flatten())
plot_decision_regions(test_X, test_y.flatten(), clf=svm, legend=2);
plt.xlim(-0.6, 0.4)
plt.ylim(-0.8, 0.8)
```

```python
# L2 regularization
from keras import regularizers

model = Sequential()
model.add(Dense(16, activation='relu', 
                kernel_regularizer=regularizers.l2(0.001),
                bias_regularizer=regularizers.l2(0.001)))
model.add(Dense(16, activation='relu', 
                kernel_regularizer=regularizers.l2(0.001), 
                bias_regularizer=regularizers.l2(0.001)))
model.add(Dense(1, activation='sigmoid', 
                kernel_regularizer=regularizers.l2(0.001), 
                bias_regularizer=regularizers.l2(0.001)))

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(train_X, train_y,epochs=200,verbose=2,validation_data=(test_X,test_y))
```

```python
#Testing accuracy and loss of neural network with L2 regularization
result = model.evaluate(test_X, test_y, verbose = 0)
print("Test score: ", result[0])
print("Test accuracy: ", result[1])
```

```python
l2regur=model.predict(test_X)
```

```python
#Dropout regularization
from keras.layers import Dropout

model = models.Sequential()
model.add(Dense(16, activation='relu',input_dim=2))
model.add(Dropout(0.3))
model.add(Dense(16,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(train_X, train_y,epochs=100,verbose=2,validation_data=(test_X,test_y))
```

```python
#Testing accuracy and loss of neural network with dropout regularization
value = model.evaluate(test_X, test_y, verbose = 0)
print("Test score: ", value[0])
print("Test accuracy: ", value[1])
```

```python
dropregur=model.predict(test_X)
```
